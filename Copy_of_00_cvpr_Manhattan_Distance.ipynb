We use **.astype(int)** because image pixels are usually stored as unsigned numbers (uint8) from 0 to 255. When you subtract two pixels, if the result is negative, it wraps around to a large positive number instead of showing the negative value. Like this:
***246-256 = -10 as expected but in uint8 it came as (256-10)=246***
Converting to int lets Python handle negative numbers correctly, so subtraction works as expected.

```python
# Manhattan Pixel wise
pixel21=image2[0,0]
pixel31=image3[0,0]
L1_distance=sum(abs(pixel21.astype(int)-pixel31.astype(int)))
L1_distance
```

```python
# Manhattan Row wise
pixel211=image2[0]
pixel311=image3[0]
L11_distance=sum(abs(pixel211.astype(int)-pixel311.astype(int)))
L11_distance
```

```python
L111_distance=0
for i in range(4000):
  L111_distance=L111_distance+ sum(abs(image2[i].astype(int)-image3[i].astype(int)))
L111_distance
```

```python
# Manhatan Total Distance for two images
manhattan_distance=np.sum(np.abs(image2.astype(int)-image3.astype(int)))
manhattan_distance
```
